````python
def get_season(column_name):
    month = int(column_name.split('.')[1])
    if 3 <= month <= 4:
        return '봄'
    elif 5 <= month <= 9:
        return '여름'
    elif 10 <= month <= 11:
        return '가을'
    else:
        return '겨울'

season_mapping = {col: get_season(col) for col in data.columns[2:28]}
df_season = data.groupby(season_mapping, axis=1).mean()
df_season
````

````python
for df_season.iloc[1].keys()
````

교양_오락_문화 | 교육 | 교통 | 내구재 | 외식 | 의류

````python
season_row={'교양_오락_문화':{'봄':0, '여름':0, '가을':0, '겨울':0},
            '교육':{'봄':0, '여름':0, '가을':0, '겨울':0},
            '교통':{'봄':0, '여름':0, '가을':0, '겨울':0},
            '내구재':{'봄':0, '여름':0, '가을':0, '겨울':0},
            '외식':{'봄':0, '여름':0, '가을':0, '겨울':0},
            '의류':{'봄':0, '여름':0, '가을':0, '겨울':0}}
for idx in df_season.index:
    if idx in [45,  46,  47,  48,  49,  50,  51]:
        for key in df_season.loc[idx].keys():
            season_row['내구재'][key]+=df_season.loc[idx][key]
    elif idx in [67,  68,  69,  70,  71,  72,  73]:
        for key in df_season.loc[idx].keys():
            season_row['의류'][key]+=df_season.loc[idx][key]
    elif idx in [89,  90,  91,  92,  93,  94,  95]:
        for key in df_season.loc[idx].keys():
            season_row['외식'][key]+=df_season.loc[idx][key]
    elif idx in [111, 112, 113, 114, 115, 116, 117]:
        for key in df_season.loc[idx].keys():
            season_row['교육'][key]+=df_season.loc[idx][key]
    elif idx in [133, 134, 135, 136, 137, 138, 139]:
        for key in df_season.loc[idx].keys():
            season_row['교양_오락_문화'][key]+=df_season.loc[idx][key]
    elif idx in [155, 156, 157, 158, 159, 160, 161]:
        for key in df_season.loc[idx].keys():
            season_row['교통'][key]+=df_season.loc[idx][key]
    
season_row
````

````python
# DataFrame 생성
season_df = pd.DataFrame(season_row).reset_index()

# '계절' 칼럼 이름 설정
season_df.rename(columns={'index': '계절'}, inplace=True)
season_df.iloc[:, 1:]=season_df.iloc[:, 1:]/7
season_df.to_csv('./data/category_weight_season.csv', encoding='utf-8-sig')
season_df
````
